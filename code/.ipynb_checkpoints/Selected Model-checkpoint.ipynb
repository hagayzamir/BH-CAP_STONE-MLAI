```python
# Import libraries
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
# from xgboost import XGBClassifier
import xgboost as xgb
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, plot_confusion_matrix
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize, TweetTokenizer
from nltk.corpus import stopwords
import nltk
import warnings
from sklearn.exceptions import ConvergenceWarning
# Suppress warnings
warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('../data/original_raw_data.csv')
```

```python
# Removing the 'Unnamed: 0' column
df = df.drop('Unnamed: 0', axis=1)
```

```python
# Initialize lemmatizer, stemmer and TweetTokenizer
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
tweet_tokenizer = TweetTokenizer()
```

```python
# Define stopwords
stop_words = set(stopwords.words('english'))

# Remove 'rt' as it refer to a re-tweet in the tweeet and is non important for our model
stop_words.update(['RT', 'I'])
```

```python
def preprocess_text(text):
    # Tokenize the text
    word_tokens = tweet_tokenizer.tokenize(text)

    # Lemmatize the tokens, stem the tokens, remove stopwords and non-alphabetic tokens
    processed_tokens = [stemmer.stem(lemmatizer.lemmatize(w)) for w in word_tokens if w not in stop_words and w.isalpha()]

    # Join the tokens back into a single string and return it
    return ' '.join(processed_tokens)
```

```python
# Apply the preprocessing function to the 'tweet' column
df['processed_tweet'] = df['tweet'].apply(preprocess_text)
df
```

```python
# Pie chart for the categorical column (class)
plt.figure()
class_counts = df['class'].value_counts()
plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of class')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.legend(class_counts.index, title="Class", loc="best")  # Add legend to the pie chart
#plt.savefig('../plots/Pie Chart of class.png')
plt.show()
```

```python
# Prepare the text data
X_processed = df['processed_tweet']
y = df['class']

X_train_processed, X_test_processed, y_train_processed, y_test_processed = train_test_split(X_processed, y, test_size=0.3, random_state=42)
```

# GridSearchCV for eXtreme Gradient Boosting

```python
# Initialize vectorizer and classifier
vectorizer = CountVectorizer()
classifier = xgb.XGBClassifier(random_state=42, reg_lambda=0.1, reg_alpha=0.0)

# Create a pipeline
pipe = Pipeline([
    ('count_vectorizer', vectorizer),
    ('xgb', classifier)
])

# Define the parameter grid for CountVectorizer and XGBClassifier
param_grid = {
    'count_vectorizer__ngram_range': [(1, 1), (1, 2)],
    'count_vectorizer__stop_words': [None, 'english'],
    'count_vectorizer__min_df': [1, 2, 5],
    'count_vectorizer__max_features': [None, 500, 1000]
}

# Perform GridSearch
grid = GridSearchCV(pipe, param_grid, cv=5, verbose=1)
start_time = time.time()
grid.fit(X_train_processed, y_train_processed)
elapsed_time = (time.time() - start_time) / 60

# Best parameters and score
print(f"Best parameters: {grid.best_params_}")
print(f"Best score: {grid.best_score_}")
print(f"Fit time: {elapsed_time} minutes")
```

```python
# Create an empty DataFrame to store the results for each model
results_df = pd.DataFrame(columns=['Model', 'Train Time', 'Accuracy', 'Precision', 'Recall', 'F1-Score', 'Support'])
```

# eXtreme Gradient Boosting

```python
# Initialize vectorizer and classifier with the optimal parameters
vectorizer = CountVectorizer(max_features=None, min_df=5, ngram_range=(1, 1), stop_words=None)
classifier = xgb.XGBClassifier(random_state=42, reg_lambda=0.1, reg_alpha=0.0)

# Create a pipeline
pipe = Pipeline([
    ('count_vectorizer', vectorizer),
    ('xgb', classifier)
])

# Start the timer
start_time = time.time()

# Fit the pipeline on your training data
pipe.fit(X_train_processed, y_train_processed)

# Stop the timer and print the elapsed time
elapsed_time = time.time() - start_time
print(f"Model trained successfully. Time elapsed: {elapsed_time} seconds.")

# Predict the test set results
y_pred = pipe.predict(X_test_processed)

# Compute and print accuracy
accuracy = accuracy_score(y_test_processed, y_pred)
print(f"Accuracy: {accuracy}")

# Compute and print confusion matrix
confusion_mat = confusion_matrix(y_test_processed, y_pred)
print(f"Confusion Matrix: \n{confusion_mat}")

# Get the classification report as a dictionary
report = classification_report(y_test_processed, y_pred, output_dict=True)

# Add the results to the DataFrame
results_df = results_df.append({
    'Model': 'eXtreme Gradient Boosting',
    'Train Time': elapsed_time,
    'Accuracy': accuracy,
    'Precision': report['weighted avg']['precision'],
    'Recall': report['weighted avg']['recall'],
    'F1-Score': report['weighted avg']['f1-score'],
    'Support': report['weighted avg']['support']
}, ignore_index=True)


# Plot confusion matrix
plot_confusion_matrix(pipe, X_test_processed, y_test_processed, cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```
