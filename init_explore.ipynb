```python
def assign_winnin_team(matches, home_team_goal, away_team_goal, winners):
    """
    Calculates a new column based on the values of two other columns in a dataframe.

    Args:
        df (pandas.DataFrame): The dataframe to add the new column to.
        column1 (str): The name of the first column to use in the calculation.
        column2 (str): The name of the second column to use in the calculation.
        new_column_name (str): The name of the new column to create.

    Returns:
        None
    """
    if matches[home_team_goal] > matches[away_team_goal]:
        matches[winners] = 'HW'
    elif matches[home_team_goal] < matches[away_team_goal]:
        matches[winners] = 'AW'
    else:
        matches[winners] = 'NW'
```
